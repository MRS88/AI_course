### Задача 1

С помощью функции input считайте с входящей строки целое положительное число и сохраните его в переменную N. Используя конструкцию if...else, выведите на экран результат в зависимости от условий:

* если N нечетное, то выведите на экран "N нечетное",
* если N четное и входит в интервал от 5 до 10 включительно, выведите на экран "N четное и принадлежит интервалу [5, 10]",
* если N четное и больше 10, выведите на экран "N четное и N > 10",
* если N четное и меньше 5, выведите на экран "N четное и N < 5".

**Решение:**

```
N = int (input("Введите целое положительное число: ")) # Просим пользователя ввести целое положительное число
if N % 2 != 0: # Проверяем, что введенное число не делится на 2 без остатка (это означает, что число нечетно, например, 5 % 2 = 1)
  print (N, " нечетное") # Выводим сообщение о том, что введенное число нечетное
elif N >= 5 and N <= 10: # Проверяем, что введенное число больше или равно 5 и меньше или равно 10 (здесь уже можно не проверять, что число четное)
  print (N, "четное и принадлежит интервалу [5, 10]") # Выводим сообщение о том, что число принадлежит интервалу [5, 10]
elif N > 10: # Проверяем, что число больше 10
   print (N, "четное и ", N, " > 10") # Выводим сообщение о том, что введенное число четное и больше 10
else: # Во всех остальных случаях
   print (N, "четное и ", N, " < 5")  # Выводим сообщение о том, что введенное число четное и меньше 5 (других вариантов уже не остается)
```

### Задача 2

Используя цикл while, посчитайте и выведите на экран сумму всех целых чисел от 15 до 22 включительно.


**Решение:**

```
x = 15 # Создаем переменную х и присваимваем ей значение 15 (стартовое значение)
summa = 0 # Задаем переменную, в которой будем хранить сумму чисел и присваиваем ей значение 0
while x < 23: # Создаем цикл while (выполняется до тех пор, пока x меньше 23)
  summa += x # Увеличиваем значение переменной summa на x (на первом шаге цикла summa += x (0 += 15) (15)) (на втором шаге цикла summa += x (15 +=16) (31)) и так далее пока x < 23
  x += 1 # Увеличиваем значение переменной x на единицу
print (summa) # Выводим посчитанную сумму чисел
```

### Задача 3

Используя цикл for, найдите сумму всех элементов заданного списка (без использования встроенных функций sum, len, sort и т. д.).

```
my_list = [56, 23, 67, 45, 67, 2, 47, 158, 31, 34, 78, 23, 78, 23, 89, 23, 36]
```

**Решение:**

```
summa = 0 # Создаем переменную summa, в которой будем хранить сумму элементов и присваиваем ей значение 0
for el in my_list: # Проходим в цикле for по всем элементам исходного списка
  summa += el # Увеличиваем значение переменной summa на текущий элемент списка (1 шаг цикла: summa += el (0 += 56) (56)) (2 шаг цикла: summa +=el (56 += 23) (79)) и так по всем элементам списка
print ("Сумма всех элементов заданного списка равна ", summa) # Выводим полученную сумму
```

### Задача 4

Используя цикл (любой), найдите значение максимального элемента списка из предыдущего задания (без использования встроенных функций max и т. д.). Можно использовать встроенные методы списков (любые). 

P.S. Не стесняйтесь гуглить :)

P.S. 2 Гуглить нужно фразу "алгоритм поиска максимального элемента массива".

Пример: для списка [3, 4, 1, 7, 2] значением максимального элемента является число 7.

**Решение:**

```
max_el = my_list[0] # Создаем переменную, в которой будем хранить значение максимального элемента списка и присваиваем ей значение первого элемента списка
for el in my_list: # Проходим по всем элементам исходного списка
  if el > max_el: # Проверяем, если текущий элемент списка больше значения max_el
    max_el = el # Записываем в переменную max_el значение текущего элемента
print ("Максимальный элемент, найденный перебором с использованием цикла for: ", max_el) # Выводим найденные максимальный элемент
```

### Задача 5

Создайте словарь, соответствующий следующему описанию: в честь 8 Марта начальник отдела Валера решил принести на работу коробку конфет и угостить коллег :) Красотка Наташа съела две конфеты, ее подруга Алина - целых три, разработчик Марат унес с собой в соседний опен-спейс целых пятнадцать, чтобы поделиться со своей командой, менеджер проекта Лев проходил мимо и съел одну конфету, а сам Валера, будучи сторонником здорового образа жизни, не съел ни одной.

Добавьте в словарь данные для студента Ромы, который работает неполный рабочий день, и, придя на работу после экзамена, с удовольствием съел 4 конфеты.

P.S. Все совпадения случайны :) Для добавления Ромы используйте встроенный метод .update

**Решение:**

```
# Можно использовать и такой вариант
my_dict = {'Наташа' : 2,  # Создаем словарь my_dict и задаем ключ 'Наташа' со значением 2
           'Алина' : 3, # задаем ключ 'Алина' со значением 3
           'Марат' : 15, # задаем ключ 'Марат' со значением 15
           'Лев' : 1, # задаем ключ 'Лев' со значением 1
           'Валера' : 0 # задаем ключ 'Валера' со значением 0
          }
print(my_dict)

my_dict.update({'Рома' : 4}) # Добавляет в словарь С помощью метода update новую пару элементв: ключ 'Рома' - значение 'Четыре конфеты'
print (my_dict) # Выводим измененные в словарь
```
