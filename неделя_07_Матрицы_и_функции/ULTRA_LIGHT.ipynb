#Задача 1

* Создайте скаляр
* Создайте вектор размера 5
* Создайте матрицу размера 5х5

Выведите все три переменные на экран, если это возможно, выведите их форму и общее число элементов

```
import numpy as np

a = 6.5
print(a, '- скаляр\n')
vec_h = np.arange(1, 6, 1).reshape(1, 5)
print(vec_h, '- горизонтальный вектор')
print(vec_h.shape, '- размер гориз.вектора')
print(vec_h.size, '- общее число элементов\n')

vec_v = vec_h.reshape(5,1)
print(vec_v, '- вертикальный вектор')
print(vec_v.shape, '- размер вертик.вектора')
print(vec_v.size, '- общее число элементов\n')

M = vec_v.dot(vec_h)
print(M, '- матрица')
print(M.shape, '- размер матрицы')
print(M.size, '- количество элементов в матрице')
```
